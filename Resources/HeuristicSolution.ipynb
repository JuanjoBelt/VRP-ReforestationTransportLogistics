# Proposed algorithm

```python
demandas = d.copy()
n0 = 'base'
ruta = [n0]
rutas = []
costos = []
while demandas.sum()["Demanda"] > 0:
    carga = 1
    ruta = [n0]
    costo = 1
    while carga > 0 and demandas["Demanda"].sum() > 0:
        n = df.loc[n0, np.array(demandas["Demanda"]>0)].idxmax()
        ruta.append(n)
        
        if carga > demandas.loc[n].Demanda:
            carga -= demandas.loc[n].Demanda
            demandas.loc[n].Demanda = 0
            costo += df.loc[ruta[-2],n]/v
            
        elif carga == demandas.loc[n].Demanda:
            carga = 0
            demandas.loc[n].Demanda = 0
            ruta.append(n0)
            costo += df.loc[ruta[-2],n0]/v
            
        else:
            demandas.loc[n].Demanda -= carga
            carga = 0
            ruta.append(n0)
            costo += df.loc[ruta[-2],n0]/v
            
        while carga > 0 and demandas["Demanda"].sum() > 0:
            if df.loc[df.loc[n, np.array(demandas["Demanda"]>0)].idxmin(),n] < df.loc[n,'base']:
                n = df.loc[n, np.array(demandas["Demanda"]>0)].idxmin()

                ruta.append(n)
                
                if carga > demandas.loc[n].Demanda:
                    carga -= demandas.loc[n].Demanda
                    demandas.loc[n].Demanda = 0
                    costo += df.loc[ruta[-2],n]/v
                    
                elif carga == demandas.loc[n].Demanda:
                    carga = 0
                    demandas.loc[n].Demanda = 0
                    ruta.append(n0)
                    costo += df.loc[ruta[-2],n0]/v
                    
                else:
                    demandas.loc[n].Demanda -= carga
                    carga = 0
                    ruta.append(n0)
                    costo += df.loc[ruta[-2],n0]/v
                    
            else:
                carga = 0
                ruta.append(n0)
                costo += df.loc[ruta[-2],n0]/v
            
        
    rutas.append(ruta)
    costos.append(costo)
if rutas[-1][-1] != n0:
    rutas[-1].append(n0)
    costo += df.loc[ruta[-2],n0]
    
    
rutas
```

```python
df = df_ruta_costos

# Create a new list to store the modified rows
new_rows = []

# Iterate over the dataframe
for index, row in df.iterrows():
    cost = row['Cost']
    route = row['Route']
    
    # Check if the integer part of the cost is even and greater than 0
    if int(cost) % 2 == 0 and int(cost) != 0:
        integer_part = int(cost)
        decimal_part = cost - integer_part
        
        # Divide the cost into equal parts based on the integer part
        divided_cost = 1 + (decimal_part / integer_part)
        
        # Append the route multiple times based on the integer part
        for _ in range(integer_part):
            new_rows.append({'Route': route, 'Cost': divided_cost})
    else:
        # If the cost is not even, retain the original row
        new_rows.append({'Route': route, 'Cost': cost})

# Convert new_rows to a DataFrame
new_df = pd.DataFrame(new_rows)

# Display the resulting dataframe
new_df
```

```python
# Save the resulting dataframe to a CSV file
new_df.to_csv('rutas_costos.csv', index=False)
```
